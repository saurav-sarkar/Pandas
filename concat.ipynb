```python
import pandas as pd
```

```python
ind_weather = pd.DataFrame({
    "city" : ["delhi","mumbai","banglore"],
    "temp" : [22,21,23],
    "humidity" : [36,35,34]
})
ind_weather
```

```python
us_weather = pd.DataFrame({
    "city" : ["miami","ohio","mexico"],
    "temp" : [22,21,23],
    "humidity" : [36,35,32]
})
us_weather
```

```python
df=pd.concat([ind_weather,us_weather])
df
```

```python
df=pd.concat([ind_weather,us_weather], ignore_index = True)
df
```

```python
df=pd.concat([ind_weather,us_weather],keys=["ind","us"],axis=1)
df
```
